# date to to_datetime

```python
dfComp
```

```python
dfComp['date'] = pd.to_datetime(dfComp['date'], dayfirst=True)
dfComp['date'].describe()
```

```python
dfComp.date[:4].dt.year, dfComp.date[:4].dt.month,dfComp.date[:4].dt.day
```

```python
dfComp = dfComp.set_index('date').asfreq('b')
```

```python
dfComp.isna().sum()
```

```python
# Fill nulls

dfComp.fillna(method = 'ffill', inplace= True)

dfComp.isna().sum()
```

```python
# keeping spx and dropping all attributes

dfComp.drop(['dax','ftse','nikkei'], axis =1 , inplace= True)
```

```python
dfComp.rename( columns={ 
    'spx':'market_value'
}, inplace= True)
```
